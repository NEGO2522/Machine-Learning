# Feature Scaling (Normalization)

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding = 'latin1')
df.head(3)
```

```python
df.isnull().sum()
```

```python
df['Amount'].fillna(df['Amount'].mean(),inplace=True)
```

```python
df.describe()
```

```python
sns.distplot(df['Amount'])
plt.show()
```

 Now we do min max scalling the following code . It Means Normalization 


```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mx = MinMaxScaler()
mx.fit(df[['Amount']])
```

```python
mx.transform(df[['Amount']])
```

```python
df['Amount_tran'] = mx.transform(df[['Amount']])
```

```python
df
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(df['Amount'])

plt.subplot(1,2,2)
plt.title('After')
sns.distplot(df['Amount_tran'])

plt.show()
```
