## Function Transformer In Python 

```python
import numpy as np 
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding = 'latin1')
df.head(3)
```

```python
df.isnull().sum()
```

```python
sns.distplot(df['Age'])
plt.show()
```

```python
q1 = df['Age'].quantile(0.25)
q3 = df['Age'].quantile(0.75)
iqr = q3-q1
```

```python
min_r = q1 - (1.5*iqr)
max_r = q3 + (1.5*iqr)
```

## Convert To Normal Distrbution we Use Fucntion Transformer 

```python
 from sklearn.preprocessing import FunctionTransformer
```

```python
ft = FunctionTransformer(func=np.log1p)
ft.fit(df[['Age']])
```

```python
df['Age_Tf'] = ft.transform(df[['Age']])
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
sns.distplot(df['Age'])
plt.title('Before')

plt.subplot(1,2,2)
sns.distplot(df['Age_Tf'])
plt.title('After')
plt.show()
```
