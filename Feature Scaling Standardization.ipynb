# Feature Scaling(Standardization)

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv',encoding ='latin1')
df.head(3)
```

```python
df.isnull().sum()
```

```python
df['Amount'].fillna(df['Amount'].mean(), inplace= True)
```

```python
sns.distplot(df['Age'])
plt.show()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(df[["Age"]])
```

```python
df['Age_ss'] = pd.DataFrame(ss.transform(df[['Age']]),columns=['x'])
```

```python
df.head(3)
```

```python
df.describe()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(df['Age_ss'])

plt.subplot(1,2,2)
plt.title('After')
sns.distplot(df['Age'])

plt.show()
```
