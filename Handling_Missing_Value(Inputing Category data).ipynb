## Mode Filling

```python
df['Amount'].mode()[0]
```

```python
df['Amount']=  df['Amount'].fillna(df['Status'].mode()[0])
```

```python
mode_value = df['Amount'].mode()

if not mode_value.empty:
    df['Amount'] = df['Amount'].fillna(mode_value[0])
else:
    print("No mode found for 'Amount' column.")
```
