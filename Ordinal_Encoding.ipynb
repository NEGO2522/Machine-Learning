# Ordinal Encoding In Machine Learning 

```python
import pandas as pd
df = pd.DataFrame({"Size":["s","m","l","xl","s", "m", "l", "s","s","l","xl","m"]})
df.head(3)
```

```python
ord_data = [["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
oe.fit(df[["Size"]])
```

```python
oe.transform(df[["Size"]])
df['Size_en'] = oe.transform(df[["Size"]])
```

```python
df.head(3)
```

## How to do the same thing using map function 

```python
ord_data1 = {'s':0,"m":1,"l":4,"xl":25}
df["Size_en_map"]= df['Size'].map(ord_data1)
```
